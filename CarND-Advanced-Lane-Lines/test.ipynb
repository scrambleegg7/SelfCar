```python
def is_prime(number):
    if number > 1:
        if number == 2:
            return True
        if number % 2 == 0:
            return False
        for current in range(3, int(np.sqrt(number) + 1), 2):
            if number % current == 0: 
                return False
        return True
    return False
```

```python
class Image:
    
    def __init__(self,x,y):
        self.x = x
        self.y = y
    
    def batch_next(self,rindex):
        
        return self.x[rindex]
```

```python
def get_primes(number):
    
    while True:
        if is_prime(number):
            yield number
        number += 1 # <<<<<<<<<<

def mygenerator(X,y):
    
    image = Image(X,y)
    r = np.random.permutation(X.shape[0])

    BATCH_SIZE=32
    while 1:
        for offset in range(0,X.shape[0],BATCH_SIZE):
            rindex = r[offset:offset+BATCH_SIZE]
            yield image.batch_next(rindex)
```

```python
X = np.random.random((200,10))
y = np.random.random(200)

mygen = mygenerator( X, y )
```
